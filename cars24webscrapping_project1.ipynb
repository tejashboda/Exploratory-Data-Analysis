```python
# importing libraries
import numpy as np
import pandas as pd
from bs4 import BeautifulSoup
import requests
from fake_headers import Headers
```

```python
# initialising variables
car_name=[]
model=[]
travelled=[]
Emi=[]
down_payment=[]
selling_price=[]
hyderabad_area=[]
x=[]

headers = Headers(os="windows", headers=True).generate()
city = ["3686&pinId=500001", "4709&pinId=560001", "5732&pinId=600001","2378&pinId=400001","2&pinId=110001","777&pinId=700001","2598&pinId=422001","6105&pinId=641001"]
fuel = ["petrol", "diesel", "cng",]
for i in city:
    for j in fuel:
       
            
        URL = f"https://www.cars24.com/buy-used-car?f=fuelType%3Ain%3A{j}&sort=bestmatch&serveWarrantyCount=true&gaId=1853873167.1699645025&listingSource=TabFilter&storeCityId={i}"
        page = requests.get(URL,headers=headers)
        #page = requests.get(URL,headers=headers)
        pagecontent = page.text
        s = BeautifulSoup(pagecontent)


        for z in s.findAll('div',attrs={'class':'_2z-Yu'}): # for in the main class
            
            c_name = z.find('h3',attrs={'class':'_2lmIw'})  # attributes classes extractions
            model_ = z.find('ul',attrs={'class':'_1hOnS'})
            travel = z.find('ul',attrs={'class':'_13yb6'})
            emi = z.find('div',attrs={'class':'_1Em-A'})
            d_payment = z.find('span',attrs={'class':'_2JX5N'})
            sp = z.find('div',attrs={'class':'_18ToE'})
            area = z.find('span',attrs={'class':'_3DYbK'})
             
            if i =='3686&pinId=500001':          # these ids are used to identity the city for which the data is extracted
                x.append('Hyderabad')
            elif i == '4709&pinId=560001':
                x.append('Bangaluru')
            elif i == '5732&pinId=600001':
                x.append('Chennai')
            elif i == '2378&pinId=400001':
                x.append('Mumbai')
            elif i == '2&pinId=110001':
                x.append('Delhi')
            elif i == '777&pinId=700001':
                x.append('Kolkata')
            elif i == '2598&pinId=422001':
                x.append('Nashik')
            elif i == '6105&pinId=641001':
                x.append('Coimbatore')  
                
            if c_name is None:             
                car_name.append(np.NaN)
            else:
                car_name.append(c_name.text)

            if model_ is None:
                model.append(np.NaN)
            else:
                model.append(model_.text)

            if travel is None:
                travelled.append(np.NaN)
            else:
                travelled.append(travel.text)

            if emi is None:
                Emi.append(np.NaN)
            else:
                Emi.append(emi.text)

            if d_payment is None:
                down_payment.append(np.NaN)
            else:
                down_payment.append(d_payment.text)

            if sp is None:
                selling_price.append(np.NaN)
            else:
                selling_price.append(sp.text)

            if area is None:
                hyderabad_area.append(np.NaN)
            else:
                hyderabad_area.append(area.text)
```
